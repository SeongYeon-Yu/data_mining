```python
#60181912 유성연
#데이터마이닝 5주차 과제
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    Score = np.random.randint(100, size=(10, 4))
    print(Score) # 원래 점수
    prob1(Score)
    prob2(Score)
    prob3(Score)
    prob4(Score)
```

```python
def prob1(Score):
    plt.scatter(Score[:,1:2],Score[:,0:1],marker='^')
    plt.xlabel('English')
    plt.ylabel('Korea')
    plt.show()
```

```python
def prob2(Score):
    plt.subplot(2,2,1)
    plt.hist(Score[:,0:1])
    plt.xlabel('korea')
    plt.ylabel('Count')

    plt.subplot(2,2,2)
    plt.hist(Score[:,1:2])
    plt.xlabel('English')
    plt.ylabel('Count')

    plt.subplot(2,2,3)
    plt.hist(Score[:,2:3])
    plt.xlabel('Math')
    plt.ylabel('Count')

    plt.subplot(2,2,4)
    plt.hist(Score[:,3:4])
    plt.xlabel('Science')
    plt.ylabel('Count')

    plt.subplots_adjust(hspace=0.5,wspace=0.3)
    plt.show()
```

```python
def prob3(Score):
    arrlist = np.zeros(4)
    for i in range(0,10,1):
        if(Score[i][0]>=90):
            arrlist[0]+=1
        elif(80<=Score[i][0]<90):
            arrlist[1]+=1
        elif(70<=Score[i][0]<80):
            arrlist[2]+=1
        else:
            arrlist[3]+=1
    label=['A','B','C','D']
    plt.pie(arrlist,labels=lab
            .el)
    plt.show()
```

```python
def prob4(Score):
    arrlist=np.array([[Score[:,0:1]],Score[:,1:2],Score[:,2:3],Score[:,3:4]])
    fig1, ax = plt.subplots(nrows=1,ncols=1, figsize=(8, 4))
    ax.set_xticklabels(['Kor', 'Eng', 'Math', 'Science'])
    plt.boxplot(arrlist)
    plt.show()
```

```python
main()
```
