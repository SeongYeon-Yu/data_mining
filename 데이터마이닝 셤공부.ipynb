```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

digits = load_digits()
X = digits.data
y= digits.target
X_train, X_test, y_train,y_test= train_test_split(X,y, test_size=0.3)
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
y_model = model.predict(X_test)
print("정확도", accuracy_score(y_test,y_model))
```

```python
map_osm = folium.Map(location=[37,126], zoom_start=13)
df = pd.read_csv('camping_loc.csv', encoding='cp949')
df.filter(['캠핑(야영)장명','캠핑(야영)장구분','위도','경도','주소'])
```

```python
for i in df.index:
    name = df.loc[i,'캠핑(야영)장명']
    lat = df.loc[i,'위도']
    lng = df.loc[i, '경도']
    
    marker = folium.Marker([lat,lng],tooltip=name)
    marker.add_to(map_osm)
map_osm
```
