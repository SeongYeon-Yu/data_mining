```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
ax=plt.axes(projection='3d')
pi = 3.141592
r = np.linspace(-200,200)
theta = np.linspace(0,2*pi)
theta, r = np.meshgrid(theta, r)
x = r * np.cos(theta)
y = r * np.sin(theta)
z = 4 * r
ax.plot_wireframe(x,y,z)
```
