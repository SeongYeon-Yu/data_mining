```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.random import rand
from matplotlib import font_manager,rc
 #-*- coding: utf-8 -*-

font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

df = pd.read_csv('2011년.txt',header=2,sep='\t',thousands = ',')
new_df=df.drop(0)
new_df=new_df[['지역','계','계.1']]
new_df.columns=['구','총인구','65세이상인구']
new_df['65세이상비율'] = new_df.apply(lambda row: (row['65세이상인구'] / row['총인구']),axis=1)
new_df
```

```python
구별노인인구=new_df.pivot_table(index='구',values='65세이상인구')
구별노인인구.plot(kind='barh',figsize=(8,8),grid=True)
```

<h2>2011~2020 서울 지역구별 1분기 65세이상인구/비율</h2>

```python
df = pd.read_csv('2011_2020.서울노인인구(지역구).txt',header=2,sep='\t',thousands = ',')
new_df=df.drop(0)
new_df=new_df[['기간','지역','계','계.1']]
new_df=new_df[new_df['지역']!='합계']
new_df=new_df[new_df['기간'].str.contains('.1/4')]
new_df.columns=['년도','구','총인구','65세이상인구']
new_df['65세이상비율'] = new_df.apply(lambda row: (row['65세이상인구'] / row['총인구']),axis=1)
new_df
```

```python
year=input(' 2011~2020 년도를 숫자만 입력하세요. ')
year_df=new_df[new_df['년도'].str.contains(year)]
노인인구비율=year_df.pivot_table(index='구',values='65세이상비율')
노인인구비율.plot(kind='barh',figsize=(8,8),grid=True,title="%s년도 노인비율"%(year))
```

```python
location=input(' 지역구를 입력하세요. ')
location_df=new_df[new_df['구'].str.contains(location)]
노인인구비율=location_df.pivot_table(index='년도',values='65세이상비율')
노인인구비율.plot(kind='barh',figsize=(8,8),grid=True,title="%s"%(location))
```

```python
region=eval(input("지역: "))
year = input("년도: ")

a=region[['행정구역', year +'년_계_총인구수',year+'년_계_연령구간인구수']]
a['65세이상 인구 비율'] = region[year+"년_계_연령구간인구수"]/region[year+'년_계_총인구수'] * 100
a
```

```python
font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
aa=a.pivot_table(index="행정구역", values=year+"년_계_총인구수")
aa.plot(kind='barh')
```

```python
a_location = input("시군구 입력: ")
a_location_ = region[region["행정구역"].str.contains(a_location)]
for i in range(2011,2021):
    a_location_[str(i)+'년65세이상비율'] = a_location_.apply(lambda row: (row[str(i)+'년_계_연령구간인구수'] / row[str(i)+'년_계_총인구수']),axis=1)
over65 = a_location_.iloc[:,-10:].T
over65.columns = ["비율"]
over65.index.name="년도"
over65.pivot_table(index = "년도", values="비율")
over65.plot(kind='barh', figsize=(10,10))
```
